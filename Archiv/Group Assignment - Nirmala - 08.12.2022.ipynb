```python
original_df['logamount'] = np.log(original_df['amount'])
print(original_df)  
```

```python
ggplot(original_df, aes(x='category', y='logamount')) + \
    geom_boxplot()+ \
    theme(figure_size=(20, 12))  # here you define the plot size
```

```python

# creating the 3 axes for the subplots with specified size and shared y-axis
fig, (ax1, ax2, ax3) = plt.subplots(ncols=3, sharey=True, figsize=(14,7))

sns.boxplot(
    data = original_df,
    x = 'daytime', 
    y = 'logamount',
    # plotting to the first axis created above
    ax = ax1
)
# set the title and axis labels for the chart
ax1.set(
    title  = "Box Plot",
    xlabel = "Daytime",
    ylabel = "amount"
)

sns.violinplot(
    data = original_df,     
    x = 'daytime', 
    y = 'logamount',
    inner = 'quartile',
    # plotting to the second axis created above
    ax = ax2 
)
# set the title and axis labels for the chart
ax2.set(
    title  = "Violin Plot",
    xlabel = "daytime",
    ylabel = "logamount"
)

sns.boxenplot(
    data = original_df,
    x = 'daytime', 
    y = 'logamount',
    # plotting to the third axis created above
    ax = ax3 
)
# set the title and axis labels for the chart
ax3.set(
    title  = "Boxen Plot",
    xlabel = "daytime",
    ylabel = "amount"
);
```
