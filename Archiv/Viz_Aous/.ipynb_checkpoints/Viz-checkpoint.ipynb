```python
import csv
import pandas as pd

df = pd.read_csv('madrid_transactions_country_3.csv', delimiter=',', skiprows=0, low_memory=False)
```

# Comparing Catagorical Variables:

```python
from plotnine import *
(
    ggplot(df)
    + facet_grid(facets="~region ")
    + aes(x="day_type", y="amount", fill="daytime")
    + labs(
        x="day_type",
        y="amount",
    )
    + geom_col()
)
```

```python
from plotnine import *
(
    ggplot(df)
    + facet_grid(facets="~daytime ")
    + aes(x="day_type", y="amount", fill="region")
    + labs(
        x="day_type",
        y="amount",
    )
    + geom_col()
)
```

```python
from plotnine import *
(
    ggplot(df)
    + facet_grid(facets="~daytime ")
    + aes(x="day_type", y="amount", fill="Income_classification ")
    + labs(
        x="day_type",
        y="amount",
    )
    + geom_col()
)
```

# Comparing Continious Variables:

```python
from plotnine import *
ggplot(df, 
       aes(x='GDP per capita (constant 2015 US$)',
           y='amount', 
           size = 'amount',
           color = 'customer_country')) + geom_point()
```
